```python
trainData = pd.read_csv("./../Training.csv")
testData = pd.read_csv('./../Testing.csv')

A=trainData
B=testData
```

```python
A.head()
```

```python
B.head()
```

```python
A = A.drop(["Unnamed: 133"],axis=1)
A.head()
```

```python
X = A.drop(["prognosis"],axis=1)
Y = A[["prognosis"]]
```

```python
X.shape
```

```python
Y.shape
```

```python
# train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.8,random_state=42, stratify=Y)
```

```python
clf1 = LogisticRegression()
clf2 = RandomForestClassifier()
clf3 = SVC(probability = True)
clf4 = XGBClassifier()
```

```python
# group / ensemble of models
estimator = [('LR', clf1) , ('RF', clf2) , ('SVC', clf3) , ('XGB', clf4)]
```

```python
# Voting Classifier with hard voting
vot_hard = VotingClassifier(estimators = estimator, voting ='hard')
vot_hard.fit(xtrain, ytrain.values.ravel())
ypred = vot_hard.predict(xtest)
```

```python
score = accuracy_score(ytest, ypred)*100
score
# print("Hard Voting Score % d" % score)
```

```python
# Voting Classifier with soft voting
vot_soft = VotingClassifier(estimators = estimator, voting ='soft')
vot_soft.fit(xtrain, ytrain.values.ravel())
y_pred_soft = vot_soft.predict(xtest)

y_pred_soft
```

```python
score = accuracy_score(ytest, y_pred_soft)*100
score
# print("Soft Voting Score % d" % score)
```
